# Agriculture & Climate Data Analytics and Visualization

Dataset Name: Indian Agriculture and Climate Dataset (1961-2018)  
Dataset Link: https://www.kaggle.com/datasets/swarooprangle/indian-agriculture-and-climate-dataset-1961-2018?resource=download   
Dataset Description: This dataset provides a comprehensive collection of agricultural and climate data for India spanning the years 1961 to 2018. It covers various crops, rainfall, and temperature statistics, offering valuable insights into the dynamics of agricultural production and climate patterns in the region.

### Representative Crops

Cereals: Rize,Wheat,maize,Barley  [High area, policy-driven, climate sensitive]  
Pulses/Legumes: Lentils,Peas,Beans_dry,Beans_Green,Soyabean  [Soil fertility + Rainfall Dependent]  
Oilseeds: Groundnut,Sesame,Safflower [Strongly Climate Affected, Lower water tolerance]  
Commercial/Plantation Crops: Sugarcane,Coffee,Coconut,Arecanuts,CashewNuts [High water or long term crops, Irrigation intensive]  
Vegetables: Potato,Tomato,Onion,Cabbage,Cauliflower,Carrot,Okra,Cucumber,Eggplant,Lettuce,Pumpkin [Seasonal + Market Driven Volatility]  
Fruits: Mangoes,Banana,Grapes,Apple,Orange,Papaya,Pineapple,Watermelon,Melon,Lemon,Pear,Peaches,Apricots,Cherries [Temperature & Rainfall Sensitivity]  
Spices: Pepper,Ginger,Garlic,Nutmeg,FennelCoriander,Chilligreen,ChilliDry [Small Area , High Economic Value] 

### Data Preprocessing

```python
import pandas as pd
import os
```

### Merging all datasets  

Here I merged all datasets present in my folder,I did this becuse i had individual crop datasets.

```python
data_folder = '/Users/admin/Desktop/Jupyter Projects/Agriculture & Climate'

all_dfs = []

for file in os.listdir(data_folder):
    if file.endswith('.csv'):
        file_path = os.path.join(data_folder,file)
        df = pd.read_csv(file_path)

        crop_name = file.replace(".csv", "").title()
        df["Crop_File"] = crop_name

        all_dfs.append(df)
        
master_df = pd.concat(all_dfs,ignore_index=True)
master_df.head(5).T
```

```python
master_df.columns
```

### Dropping columns that were merged  for rainfall and temperature

Here i dropped columns related to rainfall and temperature

```python
master_df.drop(['YEAR','JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC','ANN','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec','ANNUAL','JAN-FEB','MAR-MAY','JUN-SEP','OCT-DEC'],axis=1,inplace=True)
```

```python
final_df = pd.merge(master_df, temperature_df, on='Year', how='left')
```

# Data Visualization

### Average Yield By Crops

```python
yield_df = final_df[final_df["Element"] == "Yield"]

avg_yield = (
    yield_df
    .groupby("Item")["Value"]
    .mean()
    .sort_values(ascending=False)
    .head(7)
)

plt.figure()
avg_yield.plot(kind="bar")
plt.xlabel("Crop")
plt.ylabel("Average Yield")
plt.title("Average Yield of Representative Crops")
plt.show()
```

### Rainfal Trend

```python
rain_trend = final_df.groupby("Year")["Annual Rain"].mean()

plt.figure()
plt.plot(rain_trend.index, rain_trend.values)
plt.xlabel("Year")
plt.ylabel("Annual Rainfall in mm")
plt.title("Average Annual Rainfall Trend in mm")
plt.show()
```

### Temperature Trend

```python
temp_trend = final_df.groupby("Year")["Annual Temp"].mean()

plt.figure()
plt.plot(temp_trend.index, temp_trend.values)
plt.xlabel("Year")
plt.ylabel("Temperature (C)")
plt.title("Average Annual Temperature Trend")
plt.show()
```

### Rainfall Distribution by Season

```python
seasonal_rain = final_df[[
    "Jan-Feb Rain", "Mar-May Rain",
    "Jun-Sep Rain", "Oct-Dec Rain"
]].mean()

plt.figure()
seasonal_rain.plot(kind="bar")
plt.xlabel("Season")
plt.ylabel("Rainfall in mm")
plt.title("Seasonal Rainfall Distribution")
plt.show()
```

### Crop Category Visualization

```python
crop_groups = {
    # Cereals
    "Rice": "Cereals", "Wheat": "Cereals", "Maize": "Cereals", "Barley": "Cereals",

    # Pulses / Legumes
    "Lentils": "Pulses", "Peas": "Pulses",
    "Beans, dry": "Pulses", "Beans, green": "Pulses", "Soyabean": "Pulses",

    # Oilseeds
    "Groundnut": "Oilseeds", "Sesame": "Oilseeds", "Safflower": "Oilseeds",

    # Commercial / Plantation
    "Sugarcane": "Commercial", "Coffee": "Commercial",
    "Coconut": "Commercial", "Arecanuts": "Commercial", "Cashewnuts": "Commercial",

    # Vegetables
    "Potato": "Vegetables", "Tomato": "Vegetables", "Onion": "Vegetables",
    "Cabbage": "Vegetables", "Cauliflower": "Vegetables", "Carrot": "Vegetables",
    "Okra": "Vegetables", "Cucumber": "Vegetables",
    "Eggplant": "Vegetables", "Lettuce": "Vegetables", "Pumpkin": "Vegetables",

    # Fruits
    "Mangoes": "Fruits", "Banana": "Fruits", "Grapes": "Fruits",
    "Apple": "Fruits", "Orange": "Fruits", "Papaya": "Fruits",
    "Pineapple": "Fruits", "Watermelon": "Fruits", "Melon": "Fruits",
    "Lemon": "Fruits", "Pear": "Fruits", "Peaches": "Fruits",
    "Apricots": "Fruits", "Cherries": "Fruits",

    # Spices
    "Pepper": "Spices", "Ginger": "Spices", "Garlic": "Spices",
    "Nutmeg": "Spices", "FennelCoriander": "Spices",
    "Chillies, green": "Spices", "Chillies, dry": "Spices"
}

final_df["Crop_Category"] = final_df["Item"].map(crop_groups)
final_df = final_df.dropna(subset=["Crop_Category"])
```

### Production Trend by Crop Category

```python
import matplotlib.pyplot as plt

prod_df = final_df[final_df["Element"] == "Production"]

category_trend = (
    prod_df.groupby(["Year", "Crop_Category"])["Value"]
    .sum()
    .reset_index()
)

plt.figure()
for cat in category_trend["Crop_Category"].unique():
    subset = category_trend[category_trend["Crop_Category"] == cat]
    plt.plot(subset["Year"], subset["Value"], label=cat)

plt.xlabel("Year")
plt.ylabel("Total Production")
plt.title("Production Trend by Crop Category")
plt.legend()
plt.show()
```

### Average Yield By Category

```python
yield_df = final_df[master_df["Element"] == "Yield"]

avg_yield_cat = (
    yield_df.groupby("Crop_Category")["Value"]
    .mean()
    .sort_values(ascending=False)
)

plt.figure()
avg_yield_cat.plot(kind="bar")
plt.xlabel("Crop Category")
plt.ylabel("Average Yield")
plt.title("Average Yield Across Crop Categories")
plt.show()
```

### Rainfall Sensitivity by Crop Category

```python
plt.figure()

for cat in final_df["Crop_Category"].unique():
    subset = final_df[
        (final_df["Crop_Category"] == cat) &
        (final_df["Element"] == "Yield")
    ]
    plt.scatter(subset["Annual Rain"], subset["Value"], label=cat, alpha=0.6)

plt.xlabel("Annual Rainfall")
plt.ylabel("Yield")
plt.title("Rainfall vs Yield Across Crop Categories")
plt.legend()
plt.show()
```
